* Comparaison fraude vs non-fraude

```python
# --- Transformation logarithmique du montant ---
df['LogAmount'] = np.log1p(df['Amount'])

# --- Création de la figure avec 2 sous-graphiques côte à côte ---
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 1. Distribution globale du log(Montant)
sns.histplot(df['LogAmount'], bins=50, color='teal', ax=axes[0])
axes[0].set_title("Distribution du log(Montant)")
axes[0].set_xlabel("log(Montant + 1)")
axes[0].set_ylabel("Nombre de transactions")
axes[0].grid(True)

# 2. Comparaison fraude vs non-fraude
sns.kdeplot(df[df['Class']==0]['LogAmount'], label='Non fraude', shade=True, ax=axes[1])
sns.kdeplot(df[df['Class']==1]['LogAmount'], label='Fraude', shade=True, ax=axes[1])
axes[1].set_title("Distribution du log(Montant) par classe")
axes[1].set_xlabel("log(Montant + 1)")
axes[1].set_ylabel("Densité")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Premier pic : autour de log(Montant+1) ≈ 1-1.5 (environ 22 000 transactions)
Correspond à de très petits montants (quelques euros)

Second pic : autour de log(Montant+1) ≈ 2.5-3 (environ 17 000 transactions)
Correspond à des montants moyens (dizaines d'euros)

La majorité des transactions se concentrent entre log(Montant+1) = 1 et 5
En échelle originale, cela représente des montants entre ~2€ et ~150€

Des transactions existent jusqu'à log(Montant+1) ≈ 10
Cela correspond à des montants exceptionnellement élevés (> 20 000€)

Le montant est un indicateur partiel : il peut aider à identifier certaines fraudes (notamment aux extrêmes), mais ne suffit pas seul car il y a un chevauchement substantiel entre les deux classes. D'autres variables seront nécessaires pour un modèle de détection efficace.

* Formule de Cohen's d

```python
# ============================================
# 4. ANALYSE QUANTITATIVE des séparations
# ============================================
from scipy import stats
print("\n" + "="*80)
print("ANALYSE DE LA SÉPARATION DES CLASSES PAR COMPOSANTE")
print("="*80)

separation_metrics = []

for i in range(28):
    col = f'V{i+1}'
    
    # Données par classe
    normal = df[df['Class']==0][col]
    fraud = df[df['Class']==1][col]
    
    # Statistiques
    mean_normal = normal.mean()
    mean_fraud = fraud.mean()
    std_normal = normal.std()
    std_fraud = fraud.std()
    
    # Différence de moyennes standardisée (Cohen's d)
    pooled_std = np.sqrt((std_normal**2 + std_fraud**2) / 2)
    cohens_d = abs(mean_fraud - mean_normal) / pooled_std
    
    # Test de Mann-Whitney (non-paramétrique)
    statistic, p_value = stats.mannwhitneyu(normal, fraud, alternative='two-sided')
    
    separation_metrics.append({
        'Composante': col,
        'Moyenne_Normal': mean_normal,
        'Moyenne_Fraude': mean_fraud,
        'Diff_Moyennes': abs(mean_fraud - mean_normal),
        'Cohen_d': cohens_d,
        'P_value': p_value
    })

# Créer un DataFrame des résultats
sep_df = pd.DataFrame(separation_metrics)
sep_df = sep_df.sort_values('Cohen_d', ascending=False)

print("\nTop 10 des composantes les plus discriminantes (Cohen's d) :")
print(sep_df.head(10).to_string(index=False))

print(f"\nComposantes significatives (p < 0.001) : {(sep_df['P_value'] < 0.001).sum()}/28")
```

Les composantes ACP étant déjà décorrélées, nous n'avons pas besoin de matrice de corrélation entre V1-V28 !
